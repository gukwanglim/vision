```python
import numpy as np
import cv2
import sys
```

```python
cv2.__version__
```

```python
img = cv2.imread('../ch_1/figure/puppy.bmp', cv2.IMREAD_COLOR)        # numpy array 타입으로 들어옴

if img is None:
    print('failed')
    sys.exit()
    
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)

while True:
    keys = cv2.waitKey()
    if keys == 27 or keys == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
img = cv2.imread('../ch_1/figure/puppy.bmp', cv2.IMREAD_COLOR)        # numpy array 타입으로 들어옴

if img is None:
    print('failed')
    sys.exit()
    
    
img_resize = cv2.resize(img, (400, 300), interpolation = cv2.INTER_AREA)               # cv 일 경우 (가로, 세로)
cv2.imwrite('image_resize.png', img_resize)

    
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image', img)
cv2.imshow('image_resize', img_resize)

while True:
    keys = cv2.waitKey()
    if keys == 27 or keys == ord('q'):
        break

cv2.destroyAllWindows()
```
