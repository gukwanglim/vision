```python
import numpy as np
import cv2
import sys
import os

import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../ch_1/figure/fig/puppy.bmp')

if img is None:
    print('failed')
    sys.exit()
    
cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image', img)

while True:
    key = cv2.waitKey()
    
    if key == 27 or key == ord('q'):
        break
        
cv2.destroyAllWindows()
```

```python
img_lists = os.listdir('../ch_1/figure/images/')
# print(img_lists)

img_path = []

for i in img_lists:
    img_name = '../ch_1/figure/images/' + i
    img_path.append(img_name)
    
cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN,
                     cv2.WINDOW_FULLSCREEN)

idx = 0
while True:
    img = cv2.imread(img_path[idx])
    if img is None:
        print('image read failed')
        break
    cv2.imshow('scene', img)
    
    if cv2.waitKey(3000) == 27:
        break
        
    idx += 1
    if idx >= len(img_path):
        idx = 0
        
cv2.destroyAllWindows()
```

```python
src = cv2.imread('../ch_1/figure/fig/cow.png')
dst = cv2.imread('../ch_1/figure/fig/green.png')

src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)

_, mask = cv2.threshold(src_gray, 244, 255, cv2.THRESH_BINARY_INV)

h, w = mask.shape[:2]

crop = dst[300:300+h, 300:300+w]

cv2.copyTo(src, mask, crop)


cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.imshow('mask', mask)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print('video open failed')
    sys.exit()

w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

out = cv2.VideoWriter('webcom.avi', fourcc, fps, (w, h))

while True:
    ret, frame = cap.read()
    if not ret:
        print('image read failed')
        break
    edge = cv2.Canny(frame, 50, 150)
    edge = cv2.cvtColor(edgee, cv2.COLOR_GRAY2BGR)
    out.write(edge)
    cv2.imshow('image', frame)
    cv2.imshow('edge', edge)
    
    if cv2.waitKey(30) == 27:
        break


cap.release()
cv2.destroyAllWindows()
```

```python
## mouse event

def call_mouse(event, x, y, flags, param):
    global oldx, oldy
    
    if event == cv2.EVENT_LBUTTONDOWN:
        oldx, oldy = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if flags & cv2.EVENT_FLAG_LBUTTON:
            cv2.line(img, (oldx, oldy), (x, y), (0, 0, 255), 5, cv2.LINE_AA)
            cv2.imshow('img', img)
            oldx, oldy = x, y

img = np.ones((500, 700), np.uint8)*255

cv2.namedWindow('img')

cv2.setMouseCallback('img', call_mouse, img)

cv2.imshow('img', img)

cv2.waitKey()
cv2.destroyAllWindows()
```
