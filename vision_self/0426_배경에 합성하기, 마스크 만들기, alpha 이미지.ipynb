```python
import numpy as np
import cv2
import sys
import os
import glob

import matplotlib.pyplot as plt
```

```python
src = cv2.imread('../ch_1/figure/fig/airplane.bmp')
mask = cv2.imread('../ch_1/figure/fig/mask_plane.bmp')
dst = cv2.imread('../ch_1/figure/fig/field.bmp')

if src is None or mask is None or dst is None:
    print('failed')
    sys.exit()
    
cv2.copyTo(src, mask, dst)                        # 비행기 사진이 배경 사진으로 옮겨짐
        #  소스영상, 마스크, input                # copyTo는 모든 사진의 크기가 같아야한다.
    
dst1 = cv2.copyTo(src, mask)
    
cv2.imshow('src', src)
cv2.imshow('mask', mask)
cv2.imshow('dst', dst)
cv2.imshow('dst1', dst1)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## 마스크 만들어서 합성

src = cv2.imread('../ch_1/figure/fig/cow.png')
dst = cv2.imread('../ch_1/figure/fig/green.png')

h, w = src.shape[:2]                                       # src 영상의 크기
print(h, w)

dst_resize = cv2.resize(dst,(w, h), interpolation = cv2.INTER_AREA)        # dst 영상 크기 조정

src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)           # gray가 마스크를 만들기에 펺하다

_, mask = cv2.threshold(src_gray, 240, 255, cv2.THRESH_BINARY_INV)    # THRESH_BINARY : 0 또는 1, THRESH_BINARY_INV : 역전

cv2.copyTo(src, mask, dst_resize)

if src is None or dst is None:
    print('failed')
    sys.exit()
    
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.imshow('src_gray', src_gray)
cv2.imshow('mask', mask)
cv2.imshow('dst_resize', dst_resize)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## 배경(사이즈가 큼)에 소(작은 사이즈를 그대로) 합성하기

src = cv2.imread('../ch_1/figure/fig/cow.png')
dst = cv2.imread('../ch_1/figure/fig/green.png')

h, w = src.shape[:2]                                       # src 영상의 크기
print(h, w)

h_1, w_1 = dst.shape[:2]
print(h_1, w_1)

dst_1 = dst[425:1025, 300:900]                        # 배경의 일부를 뜯어서 가져온다


cv2.namedWindow('dst', cv2.WINDOW_NORMAL)

# dst_resize = cv2.resize(dst,(w, h), interpolation = cv2.INTER_AREA)        # dst 영상 크기 조정

src_gray = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)           # gray가 마스크를 만들기에 펺하다

_, mask = cv2.threshold(src_gray, 240, 255, cv2.THRESH_BINARY_INV)    # THRESH_BINARY : 0 또는 1, THRESH_BINARY_INV : 역전

cv2.copyTo(src, mask, dst_1)                       # 뜯어낸 배경에 소를 넣으면 원본 배경에도 소가 생긴다

if src is None or dst is None:
    print('failed')
    sys.exit()
    
# cv2.imshow('src', src)
cv2.imshow('dst', dst)
# cv2.imshow('src_gray', src_gray)
# cv2.imshow('mask', mask)
# cv2.imshow('dst_resize', dst_resize)
cv2.imshow('dst_1', dst_1)


cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## 알파채널을 이용한 마스크

src = cv2.imread('../ch_1/figure/fig/puppy.bmp')
sunglass_4 = cv2.imread('../ch_1/figure/fig/imgbin_sunglasses_1.png',
                     cv2.IMREAD_UNCHANGED)                           # sunglass에는 alpha 채널이 있음. IMREAD_UNCHANGED : 알파채널

# print(src.shape)
# print(sunglass.shape)

sunglass_4 = cv2.resize(sunglass_4, (300, 150))

sunglass = sunglass_4[:, :, 0:3]                      # 영상

mask = sunglass_4[:, :, -1]                           # 마스크

h, w = mask.shape[:2]

crop = src[120:120+h, 220:220+w]

cv2.copyTo(sunglass, mask, crop)
# crop[mask > 0] = (0, 0, 255)                 # 이렇게하면 빨간색 안경이 생겨난다.

if src is None or sunglass is None:
    print('failed')
    sys.exit()

    
cv2.imshow('src', src)
cv2.imshow('sunglass', sunglass)
cv2.imshow('mask', mask)
cv2.imshow('crop', crop)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img_ru = cv2.imread('../ch_1/figure/fig/rudolf.png')
img_bin = cv2.imread('../ch_1/figure/fig/imgbin_hat.png',
                    cv2.IMREAD_UNCHANGED)

# print(img_ru.shape)
# print(img_bin.shape)

cv2.namedWindow('ru', cv2.WINDOW_NORMAL)

img_bin = cv2.resize(img_bin, (350, 500))

img_bin_hat = img_bin[:, :, :3]
mask = img_bin[:, :, -1]

h, w = mask.shape[:2]

cut = img_ru[200:200+h, 340:340+w]

cv2.copyTo(img_bin_hat, mask, cut)

if img_ru is None or img_bin is None:
    print('failed')
    sys.exit()

cv2.imshow('ru', img_ru)
cv2.imshow('bin', img_bin_hat)
cv2.imshow('mask', mask)
cv2.imshow('cut', cut)

cv2.waitKey()
cv2.destroyAllWindows()
```
