```python
src1 = cv2.imread('../Ch_3_Basic_img_proc/fig/lenna256.bmp', 0)
src2 = np.zeros((256, 256), np.uint8)
cv2.circle(src2, (128, 128), 100, 200, -1)
cv2.circle(src2, (128, 128), 50, 50, -1)

dst1= cv2.add(src1, src2)

# addWeighted(src1, alpha, src2, beta, gamma[, dst[, dtype]]) -> dst
dst2 = cv2.addWeighted(src1, 0.5, src2, 0.5, 0.)

dst3 = cv2.subtract(src1, src2)

dst4 = cv2.absdiff(src1, src2)

cv2.imshow('src1', src1)
cv2.imshow('src2', src2)
cv2.imshow('dst1', dst1)
cv2.imshow('dst2', dst2)
cv2.imshow('dst3', dst3)
cv2.imshow('dst4', dst4)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## bitwise operation

src1 = np.zeros((256, 256), np.uint8)
cv2.rectangle(src1, (10, 10), (127, 248), 255, -1)

src2 = np.zeros((256, 256), np.uint8)
cv2.circle(src2, (128, 128), 100, 255, -1)

dst_bw_and = cv2.bitwise_and(src1, src2)
dst_bw_or = cv2.bitwise_or(src1, src2)
dst_bw_xor = cv2.bitwise_xor(src1, src2)
dst_bw_not = cv2.bitwise_not(src1)

plt.figure(figsize = (12, 6))
plt.subplot(231), plt.imshow(src1, cmap = 'gray'), plt.axis('off'), plt.title('src1')
plt.subplot(232), plt.imshow(src2, cmap = 'gray'), plt.axis('off'), plt.title('src2')
plt.subplot(233), plt.imshow(dst_bw_and, cmap = 'gray'), plt.axis('off'), plt.title('dst_bw_and')
plt.subplot(234), plt.imshow(dst_bw_or, cmap = 'gray'), plt.axis('off'), plt.title('dst_bw_or')
plt.subplot(235), plt.imshow(dst_bw_xor, cmap = 'gray'), plt.axis('off'), plt.title('dst_bw_xor')
plt.subplot(236), plt.imshow(dst_bw_not, cmap = 'gray'), plt.axis('off'), plt.title('dst_bw_not')
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src = cv2.imread('../Ch_3_Basic_img_proc/fig/lenna.bmp', 0)

## 히스토그램 뽑아내기

# calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]]) -> hist
hist = cv2.calcHist([src], [0], None, [256], [0, 256])
                                    # 0 ~ 255
# print(hist)
print(type(hist), hist.shape)

cv2.imshow('src', src)

plt.plot(hist)
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src = cv2.imread('../Ch_3_Basic_img_proc/fig/lenna.bmp', 1)

## 히스토그램 뽑아내기 (color)

# calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]]) -> hist
hist_b = cv2.calcHist([src], [0], None, [256], [0, 256])
hist_g = cv2.calcHist([src], [1], None, [256], [0, 256])
hist_r = cv2.calcHist([src], [2], None, [256], [0, 256])

cv2.imshow('src', src)

plt.plot(hist_b, c = 'b', label = 'blue')
plt.plot(hist_g, c = 'g', label = 'green')
plt.plot(hist_r, c = 'r', label = 'red')
plt.legend()
plt.show()

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
src = cv2.imread('../Ch_3_Basic_img_proc/fig/lenna.bmp', 1)

colors = ['b', 'g', 'r']
bgr_planes = cv2.split(src)
# print(src)
# print(bgr_planes)

for (p, c) in zip(bgr_planes, colors):
    hist = cv2.calcHist([p], [0], None, [256], [0, 256])
    plt.plot(hist, color = c)

plt.show()

cv2.lmshow('src', src)

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## opencv에서 hist

def getGrayHistImage(hist):
    imgHist = np.full((200, 256), 255, np.uint8)
    
    histMax = np.max(hist)
    
    for x in range(256):
        pt1 = (x, 200)
        pt2 = (x, 200 - int(hist[x, 0]*200/histMax))
        cv2.line(imgHist, pt1, pt2, 0)
        
    return imgHist


src = cv2.imread('../Ch_3_Basic_img_proc/fig/lenna.bmp', 0)

hist = cv2.calcHist([src], [0], None, [256], [0, 256])
histImg = getGrayHistImage(hist)

cv2.imshow('src', src)
cv2.imshow('histImg', histImg)

cv2.waitKey()
cv2.destroyAllWindows()
```
