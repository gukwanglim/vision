```python
img = cv2.imread('../ch_1/figure/cat.bmp', cv2.IMREAD_UNCHANGED)      # 알파 채널(마스크 채널)이 있을 경우 IMREAD_UNCHANGED 사용.

if img is None:
    print('image read failed')
    sys.exit()
    
cv2.imshow('img', img)

cv2.waitKey()             # key를 아무것이나 받겠다.
cv2.destroyAllWindows()
```

```python
img = cv2.imread('../ch_1/figure/puppy.bmp', cv2.IMREAD_COLOR)      # 이미지 크기를 1/2로 줄임 : IMREAD_REDUCED_COLOR_2

if img is None:
    print('image read failed')
    sys.exit()
    
cv2.namedWindow('image')                                 # 이름을 주겠다
cv2.imshow('image', img)

cv2.waitKey()             # key를 아무것이나 받겠다.
cv2.destroyAllWindows()
```

```python
img = cv2.imread('../ch_1/figure/PD_MRI.jpeg', cv2.IMREAD_GRAYSCALE)
print(img.shape)

if img is None:
    print('failed')
    sys.exit()

# resize(src, dsize[, dst[, fx[, fy[, interpolation]]]]) -> dst

reduced_mri = cv2.resize(img, (300, 300), interpolation = cv2.INTER_AREA)      # 이미지 크기 조절
cv2.imwrite('../ch_1/figure/save/reduced_my_brain.jpeg', reduced_mri)          # 이미지 저장
    
cv2.namedWindow('image', cv2.WINDOW_NORMAL)          # WINDOW_NORMAL : 이미지 창 크기조절(마우스로) 가능
cv2.imshow('image', img)                             # 디폴트 값은 WINDOW_AUTOSIZE 로 크기조절 불가능

cv2.waitKey()
cv2.destroyAllWindows()
```

```python
img = cv2.imread('../ch_1/figure/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('failed')
    sys.exit()
    
cv2.namedWindow('image', cv2.WINDOW_AUTOSIZE)
cv2.imshow('image', img)

cv2.resize(img, (100, 100), interpolation = cv2.INTER_AREA)
cv2.imwrite('../ch_1/figure/save/my_puppy.png', img)

# cv2.waitKey(3000)

while True:
    keys = cv2.waitKey()
    if keys == 27 or keys == 113:       # 아스키코드 27 = esc, 113 = q
        break
        
# while True:
#     keys = cv2.waitKey()
#     if keys == ord('q'):
#         break

cv2.destroyAllWindows()
```

```python
img = cv2.imread('../ch_1/figure/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('failed')
    sys.exit()
    
# cv2.imshow('image', img)

# while True:
#     keys = cv2.waitKey()
#     if keys == 27 or keys == 113:
#         break

# cv2.destroyAllWindows()


plt.figure(figsize = (8, 8))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
imgBGR = cv2.imread('../ch_1/figure/puppy.bmp', cv2.IMREAD_COLOR)

if img is None:
    print('failed')
    sys.exit()
    
# cv2.imshow('image', img)

# while True:
#     keys = cv2.waitKey()
#     if keys == 27 or keys == 113:
#         break

# cv2.destroyAllWindows()

imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)          # opencv와 plt의 색 순서(rgb)가 다르기 때문에 순서를 설정

plt.figure(figsize = (8, 8))
plt.imshow(imgRGB)
plt.axis('off')
plt.show()
```

```python
imgBGR = cv2.imread('../ch_1/figure/puppy.bmp', cv2.IMREAD_COLOR)
imgGray = cv2.imread('../ch_1/figure/puppy.bmp', cv2.IMREAD_GRAYSCALE)

if img is None:
    print('failed')
    sys.exit()
    
# cv2.imshow('image', img)

# while True:
#     keys = cv2.waitKey()
#     if keys == 27 or keys == 113:
#         break

# cv2.destroyAllWindows()

imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 3, figsize = (12, 12))
axes[0].imshow(imgRGB), axes[0].axis('off')
axes[1].imshow(imgBGR), axes[1].axis('off')
axes[2].imshow(imgGray, cmap = 'gray'), axes[2].axis('off')       # cmap = 'gray'이 없으면 색이 나옴. (가짜 칼라)
plt.show()


# plt.figure(figsize = (8, 8))
# plt.imshow(imgRGB)
# plt.axis('off')
# plt.show()
```

```python
img_my = cv2.imread('../ch_1/figure/my.jpg', cv2.IMREAD_COLOR)
img_my_gray = cv2.imread('../ch_1/figure/my.jpg', cv2.IMREAD_GRAYSCALE)

if img_my is None:
    print('failed')
    sys.exit()

img_my_RGB = cv2.cvtColor(img_my, cv2.COLOR_BGR2RGB)

fig, axes = plt.subplots(1, 3, figsize = (12, 12))

axes[0].imshow(img_my_RGB), axes[0].axis('off')
axes[1].imshow(img_my), axes[1].axis('off')
axes[2].imshow(img_my_gray, cmap = 'gray'), axes[2].axis('off')

plt.show()
    
    
plt.figure(figsize = (8, 8))
plt.axis('off')
plt.imshow(img_my)
plt.show()

img_my_my = cv2.cvtColor(img_my, cv2.COLOR_RGB2BGR)
cv2.imwrite('../ch_1/figure/save/my_my.png', img_my_my)
```
